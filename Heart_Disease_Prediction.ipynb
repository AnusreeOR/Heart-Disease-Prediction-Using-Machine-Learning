Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, classification_report
from sklearn.metrics import f1_score
```

Data Collection and Processing

```python
# loading the csv data to a Pandas DataFrame
heart_data = pd.read_csv('/content/heart_disease_data.csv')
```

1 --> Defective Heart

Splitting the Features and Target

```python
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the Data into Training data & Test Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
##Using stratify=y-To maintain the same ratio of target classes in both sets:
#If your dataset is imbalanced (for example, 90% class 0 and 10% class 1), then its possible that the training or test set might not have a similar distribution
# which can cause biased models.
#Keep the same proportion of each label (class) in both training and test sets as in the full dataset.
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
# training the LogisticRegression model with Training data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy on training data-Overall correctness
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
# Accuracy, Precision, Recall for training data
train_accuracy = accuracy_score(Y_train, X_train_prediction)
train_precision = precision_score(Y_train, X_train_prediction)#How reliable Disease predictions are
train_recall = recall_score(Y_train, X_train_prediction)#How well the model detects actual disease

print("Training Accuracy :", train_accuracy)
print("Training Precision:", train_precision)
print("Training Recall   :", train_recall)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
# Accuracy, Precision, Recall for test data
test_accuracy = accuracy_score(Y_test, X_test_prediction)
test_precision = precision_score(Y_test, X_test_prediction)
test_recall = recall_score(Y_test, X_test_prediction)

print("\nTest Accuracy :", test_accuracy)
print("Test Precision:", test_precision)
print("Test Recall   :", test_recall)
```

```python
# Confusion Matrix
cm = confusion_matrix(Y_test, X_test_prediction)
print("\nConfusion Matrix:\n", cm)
```

```python
# Calculate F1-score
f1 = f1_score(Y_test, X_test_prediction)
print("F1-Score:", f1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.ylabel('Actual Class')
plt.xlabel('Predicted Class')
plt.show()
```

The model correctly classified 23 patients without heart disease and 27 patients with heart disease.
It made 5 false alarms (predicted disease where there was none) and 6 misses (failed to detect disease).
Overall, the model achieved an accuracy of 82%, precision of 84%, recall of 82%, and an F1-score of 83%, showing balanced and reliable performance.

```python
# Calculate F1-score
f1 = f1_score(Y_test, X_test_prediction)
print("F1-Score:", f1)
#F1 = 0.83  strong balance between precision (0.84) and recall (0.82).
#This means the model is reliable and balanced, not overfitting toward either false alarms or missed detections.
```
